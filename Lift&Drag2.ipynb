```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#constants
rho = 1.1455 #kg/m3
uinf = [18.60,18.60,18.50,18.50] 
width = 0.0318 * 2 #from lego block size
dh = 0.01 #distance between each pitot tube
Ac = width*6*0.0114 #cross sectional area at greatest point
```

```python
pressures = np.zeros((4,16))

def getdata(filename):
  data = pd.read_csv(filename, skiprows=0)
  man_neg2 = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
  time = []
  j = 0
  for row in data.iterrows():
    if(j%18 > 0 and j%18 < 17):
      i = int(row[1]["Frame"])
      if(i < 16):
        man_neg2[i].append(float(row[1]["#"]))
    elif(j%18 == 0):
      time.append(float(row[1]["#"]))
    j = j + 1
  print(man_neg2[0])
  time = (np.array(time) - time[0])/1000
  pressures = []
  for i in range(0, np.shape(man_neg2)[0]):
    pressures.append(np.mean(man_neg2[i][60:]))
    plt.plot(time, man_neg2[i], label="Scanivalve " + str(i+1))
  plt.xlabel("Time (s)")
  plt.ylabel("Pressure (CmH20)")
  plt.legend( bbox_to_anchor=(1.05, 1), loc='upper left')

  return pressures
```

```python
pressures = (np.array(pressures) + 1033.227)*(98.0665)

for i in range(0, len(pressures[:])):
  print("Upstream: " + str(pressures[i][15]))
  print("Downstream: " + str(pressures[i][14]))

def inter_P(index, x):
  return (((pressures[index][15] - pressures[index][14])/0.82)*x + pressures[index][14])

inter_P(0,0.60)
```

```python
u = np.zeros((4,14))
def get_u(index, label):
  rake = pressures[index][:14]
  u = np.sqrt(np.abs((rake - inter_P(index, 0.60)))/(1*rho/2)) #assuming rake was 60cm into 82cm wind tunnel
  plt.plot(np.linspace(0,0.13,14),u, label=label)
  plt.xlabel("pitot tube height [m]")
  plt.ylabel("Vel [m/s]")
  return(u)
```

```python
u[0] = get_u(0, "Airfoil Center")
u[1] = get_u(1, "Airfoil Edge")
u[2] = get_u(2, "Brick Center")
u[3] = get_u(3, "Brick Edge")
plt.legend()
```

```python
#different method at integrating flow field to get drag force
#turns velocity colums into 2D flow field and integrates across the width and height over a control area

#ufield unneccessary
uField1 = np.zeros((14,3))
for i in range(14):
  uField1[i][1] = u[0][i]
  uField1[i][0] = u[1][i]
  uField1[i][2] = u[1][i]

uField2 = np.zeros((14,3))
for i in range(14):
  uField2[i][1] = u[2][i]
  uField2[i][0] = u[3][i]
  uField2[i][2] = u[3][i]

#flow field needed

flowField1 = np.zeros((14,3))

for i in range(14):
  flowField1[i][1] = u[0][i]*(uinf[0] - u[0][i])
  flowField1[i][0] = u[1][i]*(uinf[0] - u[1][i])
  flowField1[i][2] = u[1][i]*(uinf[0] - u[1][i])

flowField2 = np.zeros((14,3))

for i in range(14):
  flowField2[i][1] = u[2][i]*(uinf[2] - u[2][i])
  flowField2[i][0] = u[3][i]*(uinf[2] - u[3][i])
  flowField2[i][2] = u[3][i]*(uinf[2] - u[3][i])


def double_Integral(xmin, xmax, ymin, ymax, nx, ny, A):

    dS = ((xmax-xmin)/(nx-1)) * ((ymax-ymin)/(ny-1))

    A_Internal = A[1:-1, 1:-1]

    # sides: up, down, left, right
    (A_u, A_d, A_l, A_r) = (A[0, 1:-1], A[-1, 1:-1], A[1:-1, 0], A[1:-1, -1])

    # corners
    (A_ul, A_ur, A_dl, A_dr) = (A[0, 0], A[0, -1], A[-1, 0], A[-1, -1])

    return dS * (np.sum(A_Internal)\
                + 0.5 * (np.sum(A_u) + np.sum(A_d) + np.sum(A_l) + np.sum(A_r))\
                + 0.25 * (A_ul + A_ur + A_dl + A_dr))
    
x_min,x_max,n_points_x = (0,width,3)
y_min,y_max,n_points_y = (0,0.13,14)

#Drag of shape 1 from 2D velocity integration
print("Drag of Shape 1 [N]: " + str(double_Integral(x_min, x_max, y_min, y_max, n_points_x, n_points_y, flowField1)*rho))
#Drag of shape 2 from 2D velocity integration
print("Drag of shape 2 [N]: " +  str(double_Integral(x_min, x_max, y_min, y_max, n_points_x, n_points_y, flowField2)*rho))
```

```python
#second method to check integration of flow field... calc drag/width in both scenarios then use area under plot to find the total drag
Drag = np.zeros(4)

def get_Fd(index):
  #Fd = rho * width * sum((uinf - u)*u*dh)
  dFd = (uinf[index] - u[index])*u[index]
  dFd = np.trapz(dFd, dx=0.00994)
  return np.sum(dFd)

for i in range(4):
  Drag[i] = get_Fd(i)

def integrate_D(index):
  tri = 0.5*(Drag[index] - Drag[index + 1])*(0.5*width)
  box = (0.5 * width) * Drag[index + 1]
  Drag_Tot  = (tri + box) * 2
  return Drag_Tot*rho

Fdu1 = integrate_D(0)
Fdu2 = integrate_D(2)
print(Fdu1, Fdu2)
```

Both methods produce similar result

```python
#plot drag/m at the edges and center to visualize second integration method
plt.plot([-width/2,0,width/2],[Drag[3], Drag[2], Drag[3]], label="Worst Case Scenario")
plt.plot([-width/2,0,width/2],[Drag[1], Drag[0], Drag[1]], label="Best Case Scenario")
plt.legend()
plt.xlabel("Position along width [m]")
plt.ylabel("Drag Force per unit width [N/m]")
```

```python
#Enter your own calibration values here - my TA didn't calibrate, he just said to record no-load values and subract
strain_calbration = [[-12,-12],[-3,-37],[-4,-36]] #order is noshape, shape 1, shape 2

strain_data = np.zeros((3,2))
def get_strain(filename, index):
  strain = pd.read_csv(filename)
  strain_x = (strain["201127-Ch 1 ue"]) - strain_calbration[index][0]
  strain_y = strain["201127-Ch 2 ue"] - strain_calbration[index][1]
  plt.plot(strain_x, label="Drag")
  plt.plot(strain_y, label="Lift")
  plt.xlabel("time [s]")
  plt.ylabel("micro-strain []")
  plt.legend()
  return(np.mean(np.array(strain_x)), np.mean(np.array(strain_y)))
```

```python
#calculate CD
def Cd(force,index):
  return np.abs((force/(0.5*Ac*rho*uinf[index]**2)))

print("Drag Cofficient from Strain on Shape 1: " + str(Cd(Fds1,0)))
print("Drag Coefficient from Strain on Shape 1: " + str(Cd(Fds2,2)))
print("Drag Coefficient from Vel on Shape 1: " + str(Cd(Fdu1,0)))
print("Drag Coefficient from Vel on Shape 2: " + str(Cd(Fdu2,2)))
```
